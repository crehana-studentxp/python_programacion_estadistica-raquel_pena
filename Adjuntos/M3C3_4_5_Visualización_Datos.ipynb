#**Histograma**#

```python
# Importar las librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as pyplot
import seaborn as sns
```

```python
# Importar la base
df = pd.read_csv('sample_data/NYC_Airbnb.csv')
df.head()
```

```python
# Distribución del precio de Airbnb en Nueva York
sns.histplot(df['price'])
```

```python
new_df = df.loc[df['price']<800]
new_df.shape
```

```python
# Distribución del precio de Airbnb en Nueva York menores a $800 dls con Seaborn
sns.histplot(new_df['price'],bins=10)
```

```python
# Distribución del precio de Airbnb en Nueva York menores a $800 dls con Matplotlib
pyplot.hist(new_df['price'],bins=10)
```

```python
# Distribución del precio de Airbnb en Nueva York menores a $800 dls
sns.histplot(df['price'],bins=[0,30,50,80,100,250,500,750,1000,1250])
```

```python
# Distribución del precio de Airbnb en Nueva York menores a $800 dls
sns.histplot(new_df['price'],bins=50)
```

#**Boxplot**#

```python
import seaborn as sns
```

```python
# Boxplot en toda la población
sns.boxplot(x = df['neighbourhood_group'],  y = df['price'], palette = 'husl' )
pyplot.xticks(rotation = 90) # Girar 90grados los ejes x
```

```python
new_df = df.loc[df['price']<500]
new_df.shape
```

```python
#  Distribución del precio por barrio de departamentos menores a $500 dls.
sns.boxplot(x = new_df['neighbourhood_group'],  y = new_df['price'], palette = 'husl' )
pyplot.xticks(rotation = 90) # Girar 90grados los ejes x
```

Barrio de Manthattan

```python
# Manhattan 
mh = new_df[(new_df.neighbourhood_group == "Manhattan")]

# 25% de la muestra...
mh['price'].quantile(0.25)
```

```python
# Mediana o 50% de la muestra...
mh['price'].median()
```

```python
# 75% de la muestra...
mh['price'].quantile(0.75)
```

#**Scatter Plot**#

## Precios de casas en Seattle, Estados Unidos. 

casita.jpg

```python
# Importar las librerías

import pandas as pd
import numpy as np
import matplotlib.pyplot as pyplot
import seaborn as sns
```

```python
# Importar la base
df = pd.read_csv('sample_data/kc_house_data.csv')
df.head()
```

```python
df.shape
```

Cálculo de correlaciones de todas las variables numéricas con el Precio por Pandas. El método por default: Pearson

```python
df.corr()['price'].sort_values()
```

```python
#  Distribución del precio de las casas
sns.boxplot( df['price'], color='purple' )
```

```python
# Muestra con el precio de casas menores a $800,000 dls para evitar outliers
new_df = df.loc[df['price']<800000]
new_df.shape
```

```python
# Correlación entre el precio de casa y el número de baños (matplotlib)
pyplot.scatter( x = new_df['price'], y = new_df['bathrooms'], color = 'coral')
```

```python
# Correlación entre el precio de casa y los pies cuadrados (Seaborn)

sns.scatterplot( x = new_df['price'], y = new_df['sqft_living'], color = 'deeppink')
```

```python
# Correlación entre el precio de casa y los pies cuadrados aparte del sótano(Seaborn)

sns.scatterplot( x = new_df['price'], y = new_df['sqft_above'], color = 'green')
```

Finalmente concluimos que sí existe relación positiva entre el precio de las casas y los pies cuadrados.
